```python
df['parada_fecha_hora'] = pd.to_datetime(df['parada_fecha'].astype(str) + ' ' + df['parada_hora'].astype(str), errors='coerce')
df['resolucion_fecha_hora'] = pd.to_datetime(df['resolucion_fecha'].astype(str) + ' ' + df['resolucion_hora'].astype(str), errors='coerce')

# Filtrar para la línea L43 y la máquina llenadora
df_l43_llenadora = df[(df['linea'].str.lower() == 'l43') & (df['maquina'].str.lower() == 'llenadora')]

# Ordenar los datos por fecha y hora de parada
df_l43_llenadora = df_l43_llenadora.sort_values(by='parada_fecha_hora')

# Calcular los tiempos entre fallas
df_l43_llenadora['tiempo_entre_fallas'] = df_l43_llenadora['parada_fecha_hora'].shift(-1) - df_l43_llenadora['resolucion_fecha_hora']
df_l43_llenadora['tiempo_entre_fallas'] = df_l43_llenadora['tiempo_entre_fallas'].dt.total_seconds() / 60  # en minutos

# Eliminar la última fila porque tendrá un NaN en tiempo_entre_fallas
df_l43_llenadora = df_l43_llenadora.dropna(subset=['tiempo_entre_fallas'])

# Filtrar los tiempos entre fallas para asegurarse de que son mayores que cero
df_l43_llenadora = df_l43_llenadora[df_l43_llenadora['tiempo_entre_fallas'] > 0]

# Visualización inicial de los tiempos entre fallas
sns.boxplot(df_l43_llenadora['tiempo_entre_fallas'])
plt.show()

# Filtrar valores atípicos
Q1 = df_l43_llenadora['tiempo_entre_fallas'].quantile(0.25)
Q3 = df_l43_llenadora['tiempo_entre_fallas'].quantile(0.75)
IQR = Q3 - Q1
filter = (df_l43_llenadora['tiempo_entre_fallas'] >= Q1 - 1.5 * IQR) & (df_l43_llenadora['tiempo_entre_fallas'] <= Q3 + 1.5 * IQR)
df_l43_llenadora_filtered = df_l43_llenadora.loc[filter]

# Convertir a array de NumPy
tiempos_entre_fallas_array_filtered = df_l43_llenadora_filtered['tiempo_entre_fallas'].to_numpy()

# Dividir los datos en conjuntos de entrenamiento y prueba
train_data, test_data = train_test_split(tiempos_entre_fallas_array_filtered, test_size=0.2, random_state=42)

# Normalizar los datos
mean = train_data.mean()
std = train_data.std()
train_data_norm = (train_data - mean) / std
test_data_norm = (test_data - mean) / std

# Convertir a formato adecuado para Keras
train_data_norm = train_data_norm.reshape(-1, 1)
test_data_norm = test_data_norm.reshape(-1, 1)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Definir el modelo
model = Sequential()
model.add(Dense(50, input_dim=1, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dropout(0.2))  # Añadir Dropout para regularización
model.add(Dense(1, activation='linear'))

# Compilar el modelo
model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error')

# Entrenar el modelo
history = model.fit(train_data_norm, train_data, epochs=200, batch_size=10, validation_data=(test_data_norm, test_data))

# Evaluar el modelo en los datos de prueba
test_loss = model.evaluate(test_data_norm, test_data)
print("Test Loss:", test_loss)

# Predicción y visualización
predictions = model.predict(test_data_norm)

# Graficar resultados
plt.figure(figsize=(10, 6))
plt.plot(test_data, label='True Values')
plt.plot(predictions, label='Predicted Values')
plt.legend()
plt.show()

# Graficar la pérdida durante el entrenamiento
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from scikeras.wrappers import KerasRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam, SGD
from sklearn.metrics import make_scorer, mean_squared_error

# Normalizar la característica 'tiempo_entre_fallas'
mean = df_l43_llenadora['tiempo_entre_fallas'].mean()
std = df_l43_llenadora['tiempo_entre_fallas'].std()
df_l43_llenadora['tiempo_entre_fallas'] = (df_l43_llenadora['tiempo_entre_fallas'] - mean) / std
features = df_l43_llenadora[['tiempo_entre_fallas']]

# Dividir en conjuntos de entrenamiento y prueba
train_features, test_features, train_target, test_target = train_test_split(
    features, df_l43_llenadora['tiempo_entre_fallas'], test_size=0.2, random_state=42)

# Definir el modelo para KerasRegressor
def build_model(optimizer='adam'):
    model = Sequential()
    model.add(Dense(50, input_dim=train_features.shape[1], activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1, activation='linear'))
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    return model

# Envolver el modelo de Keras en un KerasRegressor
model = KerasRegressor(model=build_model, verbose=0)

# Definir los parámetros de la búsqueda en cuadrícula
param_grid = {
    'batch_size': [10, 20, 40],
    'epochs': [100, 200],
    'model__optimizer': ['sgd', 'adam']
}

# Crear un make_scorer
scorer = make_scorer(mean_squared_error, greater_is_better=False)

# Configurar y ejecutar la búsqueda en cuadrícula
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3, scoring=scorer, error_score='raise')
grid_result = grid.fit(train_features, train_target)

# Mostrar los mejores parámetros encontrados
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))

# Evaluar el mejor modelo en el conjunto de prueba
best_model = grid_result.best_estimator_
test_loss = mean_squared_error(test_target, best_model.predict(test_features))
print("Test Loss:", test_loss)

# Predicción y visualización
predictions = best_model.predict(test_features)

# Graficar resultados
plt.figure(figsize=(10, 6))
plt.plot(test_target.values, label='True Values')
plt.plot(predictions, label='Predicted Values')
plt.legend()
plt.show()
```
